### 10 häufigste Auflösungen

### Häufigkeiten von Proportionen

```python
sizes_array = np.array(sizes)
p = np.divide(sizes_array.take(0, axis=1), sizes_array.take(1, axis=1))
proporsitions, counts = np.unique(p, return_counts=True)
```

```python
i = proporsitions[np.argsort(-counts)]
co = counts[np.argsort(-counts)]
print(len(proporsitions))
for s, c in zip(i, co):
    print(s, ':', c)
```
